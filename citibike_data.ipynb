```python
# Import dependencies
import pandas as pd
import numpy as np
```

```python
# Create necessary lists to bring in data for dataframe creation.
monthly_datasets = ['Data/JC-202301-citibike-tripdata.csv', 'Data/JC-202302-citibike-tripdata.csv', 'Data/JC-202303-citibike-tripdata.csv', 'Data/JC-202304-citibike-tripdata.csv',
                    'Data/JC-202305-citibike-tripdata.csv']

citi2023 = []
```

```python
#Loop through data and create dataframe
for month in monthly_datasets:
    month23 = pd.read_csv(month)
    citi2023.append(month23)
```

```python
# Concatenate the lists together to make the dataframe.
citibike2023 = pd.concat(citi2023)
citibike2023.info()
```

```python
# Approximate radius of earth in km
R = 6373.0

lat1 = np.radians(citibike2023["start_lat"])
lat2 = np.radians(citibike2023["end_lat"])
lng1 = np.radians(citibike2023["start_lng"])
lng2 = np.radians(citibike2023["end_lng"])
dlng = lng2 - lng1
dlat = lat2 - lat1

a = np.sin(dlat / 2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlng / 2)**2
c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1 - a))

citibike2023["distance"] = R*c

```

```python
citibike2023.to_csv('Data/citibike2023.csv')
```
